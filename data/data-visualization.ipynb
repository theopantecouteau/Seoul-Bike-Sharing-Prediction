```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('./data/seoul-bike-data-clean-for-dataviz.csv')
```

```python
df.describe().T
```

```python
# 1. Rented Bike Count by Hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Hour", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 1.Bis Rented Bike Count by Hour
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='Rented Bike Count', data=df)
plt.title('Bike Rentals Distribution by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Rented Bike Count')
plt.show()
```

```python
# 1. Rented Bike Count by Day
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Day", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Hour")
plt.xlabel("Day")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 1. Rented Bike Count by Month
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Month", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Hour")
plt.xlabel("Month")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 2. Average Temperature vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Temperature(°C)", y="Rented Bike Count")
plt.title("Average Temperature vs Rented Bike Count")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 3. Rented Bike Count by Season
df.groupby('Seasons').sum()['Rented Bike Count'].plot.pie()
```

```python
# 4. Rented Bike Count by Holiday
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Holiday", y="Rented Bike Count")
plt.title("Rented Bike Count by Holiday")
plt.xlabel("Holiday (0: No, 1: Yes)")
plt.ylabel("Rented Bike Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="WeekDay", y="Rented Bike Count")
plt.title("Rented Bike Count by Holiday")
plt.ylabel("Rented Bike Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Rainfall(mm)", y="Rented Bike Count")
plt.title("Rented Bike Count by Holiday")
plt.xlabel("Holiday (0: No, 1: Yes)")
plt.ylabel("Rented Bike Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Snowfall (cm)", y="Rented Bike Count")
plt.title("Rented Bike Count by Holiday")
plt.xlabel("Holiday (0: No, 1: Yes)")
plt.ylabel("Rented Bike Count")
plt.show()
```
