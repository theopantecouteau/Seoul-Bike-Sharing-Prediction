```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('./data/seoul-bike-data-clean.csv')
```

```python
df.describe().T
```

```python
# 1. Rented Bike Count by Hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Hour", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 1.Bis Rented Bike Count by Hour
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='Rented Bike Count', data=df)
plt.title('Bike Rentals Distribution by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Rented Bike Count')
plt.show()
```

```python
# 2. Average Temperature vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Temperature(°C)", y="Rented Bike Count")
plt.title("Average Temperature vs Rented Bike Count")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 3. Rented Bike Count by Season
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Seasons", y="Rented Bike Count")
plt.title("Rented Bike Count by Season")
plt.xlabel("Season")
plt.ylabel("Rented Bike Count")
plt.show()
```

```python
# 4. Rented Bike Count by Holiday
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Holiday", y="Rented Bike Count")
plt.title("Rented Bike Count by Holiday")
plt.xlabel("Holiday (0: No, 1: Yes)")
plt.ylabel("Rented Bike Count")
plt.show()
```

```python
# 5. Wind Speed vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Wind speed (m/s)", y="Rented Bike Count")
plt.title("Wind Speed vs Rented Bike Count")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 6. Humidity vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Humidity(%)", y="Rented Bike Count")
plt.title("Humidity vs Rented Bike Count")
plt.xlabel("Humidity (%)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 7. Solar Radiation vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Solar Radiation (MJ/m2)", y="Rented Bike Count")
plt.title("Solar Radiation vs Rented Bike Count")
plt.xlabel("Solar Radiation (MJ/m2)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 8. Visibility vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Visibility (10m)", y="Rented Bike Count")
plt.title("Visibility vs Rented Bike Count")
plt.xlabel("Visibility (10m)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 9. Rainfall vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Rainfall(mm)", y="Rented Bike Count")
plt.title("Rainfall vs Rented Bike Count")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 10. Snowfall vs Rented Bike Count
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Snowfall (cm)", y="Rented Bike Count")
plt.title("Snowfall vs Rented Bike Count")
plt.xlabel("Snowfall (cm)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 11. Rented Bike Count Over Time
plt.figure(figsize=(15, 6))
sns.lineplot(data=df, x="Date", y="Rented Bike Count")
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.title("Rented Bike Count Over Time")
plt.xlabel("Date")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

```python
# 12. Average Daily Temperature Over Time
plt.figure(figsize=(15, 6))
df_daily = df.groupby(df['Date']).mean()
sns.lineplot(data=df_daily, x=df_daily.index, y="Temperature(°C)")
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.title("Average Daily Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Average Temperature (°C)")
plt.grid(True)
plt.show()
```
