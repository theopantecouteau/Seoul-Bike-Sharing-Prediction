```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/seoul-bike-data-clean-for-dataviz.csv')
df.drop(columns="Unnamed: 0", inplace=True)
```

```python
df.describe().T
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Hour", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

We can see on the graph above that the rented bike count is at its lowest at 4am and at its highest at 6pm, with a local maximum at 8am and a local minimum at 2pm.
In fact, the rented bike count is as its highest during the commuting hours, which is not surprising.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='Rented Bike Count', data=df)
plt.title('Bike Rentals Distribution by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Rented Bike Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Day", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Day")
plt.xlabel("Day")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

No particular trend can be seen on the graph above.

```python
# 1. Rented Bike Count by Month
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Month", y="Rented Bike Count", markers='o')
plt.title("Rented Bike Count by Month")
plt.xlabel("Month")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

We can see that the rented bike count is increasing from January to June, then decreasing from July to December.
It seems consistent with the fact that the weather is getting warmer from January to June, then colder from July to December.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Temperature(°C)", y="Rented Bike Count")
plt.title("Average Temperature vs Temperature(°C)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rented Bike Count")
plt.grid(True)
plt.show()
```

As temperatures rise, bike rentals increase, suggesting that warmer weather encourages more people to rent bikes until 33°C, then the rented bike count decreases.

```python
df.groupby('Seasons').sum()['Rented Bike Count'].plot.pie(autopct='%1.1f%%')
plt.title("Rented Bike Count by Season")
plt.ylabel("Rented Bike Count")
plt.show()
```

This pie chart demonstrates that bike rentals are fairly evenly distributed across the seasons, apart from Winter which is less popular.

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Holiday", y="Rented Bike Count")
plt.title("Rented Bike Count by Holiday")
plt.xlabel("Holiday ?")
plt.ylabel("Rented Bike Count")
plt.show()
```

This bar graph appears to show that bike rental counts do not significantly differ between holidays and non-holidays, indicating consistent usage.


```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="WeekDay", y="Rented Bike Count")
plt.title("Rented Bike Count by week day")
plt.ylabel("Rented Bike Count")
plt.show()
```

This bar graph appears to show that bike rental counts do not significantly differ between week days, indicating consistent usage.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Rainfall(mm)", y="Rented Bike Count")
plt.title("Rented Bike Count by Rainfall")
plt.xlabel("Rainfall")
plt.ylabel("Rented Bike Count")
plt.show()
```

Higher rainfall seems to correlate with a decrease in bike rentals, implying adverse weather conditions deter bike usage.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Snowfall (cm)", y="Rented Bike Count")
plt.title("Rented Bike Count by Snowfall")
plt.ylabel("Rented Bike Count")
plt.show()
```

 The presence of snowfall also seems to have a negative correlation with bike rentals, indicating inclement weather likely reduces the inclination or ability to rent bikes.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Humidity(%)", y="Rented Bike Count")
plt.title("Rented Bike Count by Humidity")
plt.xlabel("Humidity")
plt.ylabel("Rented Bike Count")
plt.show()
```

The graph above shows that the rented bike count is at its lowest when the humidity is at its highest, and at its highest when the humidity is at its lowest.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Wind speed (m/s)", y="Rented Bike Count")
plt.title("Rented Bike Count by Wind speed")
plt.xlabel("Wind speed")
plt.ylabel("Rented Bike Count")
plt.show()
```

The graph above shows that the rented bike count is at its lowest when the wind speed is at its highest, and at its highest when the wind speed is at its lowest.
The drastic increase in rented bike count around 7 m/S is probably due to a special day

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Visibility (10m)", y="Rented Bike Count")
plt.title("Rented Bike Count by Visibility")
plt.xlabel("Visibility")
plt.ylabel("Rented Bike Count")
plt.show()
```

There is a pretty consistent distribution of rented bike count across the visibility range

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Solar Radiation (MJ/m2)", y="Rented Bike Count")
plt.title("Rented Bike Count by Solar Radiation")
plt.xlabel("Solar Radiation")
plt.ylabel("Rented Bike Count")
plt.show()
```

In the same way, there is a pretty consistent distribution of rented bike count across the solar radiation range

```python
#loop on the column and display the boxplot
for col in df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=col, data=df)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.show()
```

This boxplot is interesting in our outlier analysis. We can see on some columns that there are some outliers, but they are not so far from the median. This confirms that our data is clean and our decision to keep the outliers was a good one.
